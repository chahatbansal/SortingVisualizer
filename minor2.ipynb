```python
# libraries
import pygame
import random



# class for the drop down menu


class DropDown():
    
    # setting the parameters for the menu
    
    def __init__(self, color_menu, color_option, x, y, w, h, font, main, options):
        self.color_menu = color_menu
        self.color_option = color_option
        self.rect = pygame.Rect(x, y, w, h)
        self.font = font
        self.main = main
        self.options = options
        self.draw_menu = False
        self.menu_active = False
        self.active_option = -1
    
    # method to draw the entire menu
    
    def draw(self, WIN):
        pygame.draw.rect(WIN, self.color_menu[self.menu_active], self.rect, 0)
        msg = self.font.render(self.main, 1, (0, 0, 0))
        WIN.blit(msg, msg.get_rect(center = self.rect.center))

        if self.draw_menu:
            for i, text in enumerate(self.options):
                rect = self.rect.copy()
                rect.x += (i+1) * self.rect.width
                pygame.draw.rect(WIN, self.color_option[1 if i == self.active_option else 0], rect, 0)
                msg = self.font.render(text, 1, (0, 0, 0))
                WIN.blit(msg, msg.get_rect(center = rect.center))
    
    # method to get the current update of the menu and make necessary changes 
    # while setting the options
    
    def update(self, event_list):
        # initialize the menu and options
        mpos = pygame.mouse.get_pos()
        self.menu_active = self.rect.collidepoint(mpos)
        
        self.active_option = -1
        
        # check the option on which one was selected
        for i in range(len(self.options)):
            rect = self.rect.copy()
            rect.x += (i+1) * self.rect.width
            if rect.collidepoint(mpos):
                self.active_option = i
                break
                
        # if mouse goes out of the menu, close it
        if not self.menu_active and self.active_option == -1:
            self.draw_menu = False

        for event in event_list:
            if event.type == pygame.MOUSEBUTTONDOWN and event.button == 1:
                if self.menu_active:
                    self.draw_menu = not self.draw_menu
                elif self.draw_menu and self.active_option >= 0:
                    self.draw_menu = False
                    return self.active_option
        return -1




# initialize pygame
pygame.init()

# functions to use

# for making random array
def makearray(numberOfLines):
    # height of each bar (data to be sorted) 
    height = [0]*numberOfLines
    # colors of the lines
    colorOfLines = [colors[0]]*len(height)
    # randomised heights
    for i in range(0,numberOfLines):
        height[i]=random.randrange(100, 470)   
    return height, colorOfLines
    
# for  the width of the lines
def getwidth(height):
    if numberOfLines == 50:
        width = 14
    elif numberOfLines == 100:
        width = 8
    else:
        width = 5
    return width


# method to show the list of height 
def draw_lines(height, width, colorOfLines):
    # loop to iterate each item of list
    for i in range(1,len(height)+1): 
        # drawing each bar with respective gap 
        pygame.draw.rect(WIN, (colorOfLines[i-1]),
                         (((((WIDTH-40)/len(height))*i)+10, 40, width, height[i-1])))

        
        
# updates on the window  
def display():
    global first
    # draw the background only once to save memory
    if first == True:
        # background color
        WIN.fill((80, 110, 100))
        
        first = False
    # upper rect130,180,15
    WIN.fill((0,0,0),rect=[10, 10, 1060, 520])
    # lower rect80,150,150
    WIN.fill((0,0,0), rect=[10,540,1060,170])
    # for the reset button
    pygame.draw.rect(WIN, (100, 80, 255), reset_button)
    font1 = pygame.font.SysFont(None, 40)
    text2 = font1.render('Reset', True, (0,0,0))
    WIN.blit(text2, (858, 665))
    # for the start button
    pygame.draw.rect(WIN, (100, 80, 255), start_button)
    text1 = font1.render('Start', True, (0,0,0))
    WIN.blit(text1, (858, 568))
    # the main name
    font2 = pygame.font.SysFont('chalkduster.ttf', 140)
    text3 = font2.render('Visual-Algo', True, (123,170,20))
    WIN.blit(text3, (220, 570))
    # the menu's
    list1.draw(WIN)
    list2.draw(WIN)
    list3.draw(WIN)
    draw_lines(height, width, colorOfLines) 
    
    pygame.display.flip() 
    
    # to manage the speed of the application
    if timer == 1:
        pygame.time.delay(80)
    if timer == 2:
        pygame.time.delay(50)
    if timer == 3:
        pygame.time.delay(10)

        
def start_algo(algo):
    if algo == 0:
        bubble_sort(height, colorOfLines)
    if algo == 1:
        insertion_sort(height, colorOfLines)
    if algo == 2:
        selection_sort(height, colorOfLines)
    if algo == 3:
        merge_sort(height, 0, len(height)-1, colorOfLines)
                

        
# algorithms
        
        
# insertion sort
        
def insertion_sort(height, colorOfLines): 
    for i in range(1, len(height)): 
        pygame.event.pump()  
        display() 
        # temp key
        key = height[i] 
        # main line
        colorOfLines[i] = colors[3] 
        j = i - 1
        # if any line before main line is smaller then swap it
        while j >= 0 and key < height[j]:
            colorOfLines[j] = colors[1] 
            height[j + 1] = height[j] 
            display() 
            colorOfLines[j] = colors[2] 
            j = j - 1 
        # next line to be set as key
        height[j + 1] = key  
        display() 
        # make the original main line again white
        colorOfLines[i] = colors[2]

#bubble sort      

def bubble_sort(height, colorOfLines):
    for i in range(len(height) - 1):
        pygame.event.pump()
        # after this iteration max element will come at last 
        for j in range(len(height) - i - 1): 
            # starting is greater then next element 
            if height[j] > height[j + 1]: 
                colorOfLines[j+1] = colors[1]
                # save it in temporary variable 
                # and swap them using temporary variable
                height[j], height[j+1] = height[j+1], height[j]
#                 temp = height[j] 
#                 height[j] = height[j + 1] 
#                 height[j + 1] = temp
                display()
                colorOfLines[j+1] = colors[0]
        colorOfLines[j+1] = colors[2]
    # make the original line to green as a sign of completion of the array
    colorOfLines[0] = colors[2]
    display()

# merge sort

def merge_sort(height, l, r, colorOfLines): 
    mid = (l + r)//2
    if l < r: 
        mergesort(height, l, mid, colorOfLines) 
        mergesort(height, mid + 1, r, colorOfLines) 
        merge(height, l, mid,  mid + 1, r, colorOfLines) 

def merge(array, x1, y1, x2, y2, colorOfLines): 
    i = x1 
    j = x2 
    temp = [] 
    pygame.event.pump()  
    while i <= y1 and j <= y2: 
        colorOfLines[i] = colors[3] 
        colorOfLines[j] = colors[3] 
        display() 
        colorOfLines[i] = colors[0] 
        colorOfLines[j] = colors[0] 
        if height[i] < array[j]: 
                temp.append(height[i]) 
                i += 1
        else: 
                temp.append(height[j]) 
                j += 1
    while i <= y1: 
        colorOfLines[i] = colors[1] 
        display() 
        colorOfLines[i] = colors[0] 
        temp.append(height[i]) 
        i += 1
    while j <= y2: 
        colorOfLines[j] = colors[1] 
        display() 
        colorOfLines[j] = colors[0] 
        temp.append(height[j]) 
        j += 1
    j = 0    
    for i in range(x1, y2 + 1):  
        pygame.event.pump()  
        height[i] = temp[j] 
        j += 1
        colorOfLines[i] = colors[2] 
        display() 
        if y2 - x1 == len(height) - 2: 
            colorOfLines[i] = colors[3] 
        else:  
            colorOfLines[i] = colors[0] 


# selection sort

def selection_sort(height, colorOfLines):
    for i in range(len(height)):
        pygame.event.pump()
        colorOfLines[i] = (0,0,0)
        display()
        # We start with the first element
        lowest = i
        # This loop iterates over the unsorted items
        for j in range(i + 1, len(height)):
            colorOfLines[j] = colors[1]
            if height[j] < height[lowest]:
                lowest = j
            colorOfLines[lowest] = colors[3]
            display()
            colorOfLines[j] = colors[0]
            colorOfLines[lowest] = colors[0]
            
        # Swap values of the lowest unsorted element with the first unsorted
        # element
        height[i], height[lowest] = height[lowest], height[i]
        colorOfLines[i] = colors[2]
#         display()



# attributes for window
WIDTH, HEIGHT = 1080, 720
WIN = pygame.display.set_mode((WIDTH, HEIGHT))
pygame.display.set_caption("visualizer")



#  variabes


# main game
run = True
clock = pygame.time.Clock()
FPS = 120
default_font = pygame.font.SysFont(None, 30)
# array size
sizeoptions = [50,100,150]
option = 1
height = []
width = 0
# colors
colorOfLines = []
colors = [(255,255,255), (255,0,0), (0,255,0), (0,0,255)]
numberOfLines = 50
height, colorOfLines = makearray(numberOfLines)
# some boolean
first = True

# drop down menu
COLOR_INACTIVE = (100, 80, 255)
COLOR_ACTIVE = (100, 200, 255)
COLOR_LIST_INACTIVE = (255, 100, 100)
COLOR_LIST_ACTIVE = (255, 150, 150)
timer = 0
algo = -1

# for size options
list1 = DropDown(
    [COLOR_INACTIVE, COLOR_ACTIVE],
    [COLOR_LIST_INACTIVE, COLOR_LIST_ACTIVE],
    20, 560, 180, 40, 
    pygame.font.SysFont(None, 30), 
    "Select Size", ["low","mediun", "large"])

# for algorithm options
list2 = DropDown(
    [COLOR_INACTIVE, COLOR_ACTIVE],
    [COLOR_LIST_INACTIVE, COLOR_LIST_ACTIVE],
    20, 610, 180, 40, 
    pygame.font.SysFont(None, 30), 
    "Select Algorithm", ["Bubble-Sort","Insertion-Sort", "Selection-Sort", "Merge Sort"])

# for time options
list3 = DropDown(
    [COLOR_INACTIVE, COLOR_ACTIVE],
    [COLOR_LIST_INACTIVE, COLOR_LIST_ACTIVE],
    20, 660, 180, 40, 
    pygame.font.SysFont(None, 30), 
    "Select Speed", ["Slow", "Medium", "Fast",])

# for the buttons for reset and start
start_button = pygame.Rect(800, 560, 180, 40)
reset_button = pygame.Rect(800, 660, 180, 40) 
    

# main loop
while run:
    
    # add the buffer for fps
    clock.tick(FPS)
         
    # for the mouse click
    click = False

    # events in the screen
    event_list = pygame.event.get()
    for event in event_list:
        
        # quit the game
        if event.type == pygame.QUIT:
            run = False
        
        # get the option selected for the size of the array
        
        selected_option_size = list1.update(event_list)
        if selected_option_size >= 0:
            # change the title of the option
            list1.main = list1.options[selected_option_size]
            option = selected_option_size
            numberOfLines = sizeoptions[option]
            height, colorOfLines = makearray(numberOfLines)
        
        
        # to get the option selected for the algorithm
        
        selected_option_algo = list2.update(event_list)
        if selected_option_algo >= 0:
            # change the title of the option 
            list2.main = list2.options[selected_option_algo]
            # the algo selected to be called 
            if selected_option_algo == 0:
                algo = 0
            if selected_option_algo == 1:
                algo = 1
            if selected_option_algo == 2:
                algo = 2
            if selected_option_algo == 3:
                algo = 3            

                
       
        # get the option selected for the speed of sorting
        
        selected_option_speed = list3.update(event_list)
        if selected_option_speed >= 0:
            # change the title of the option
            list3.main = list3.options[selected_option_speed]
            # the speed selected 
            if selected_option_speed == 0:
                timer = 1
            if selected_option_speed == 1:
                timer = 2
            if selected_option_speed == 2:
                timer = 3
        
        
        #for reset button and start button
        
        mouse = pygame.mouse.get_pos()
        if event.type == pygame.MOUSEBUTTONDOWN:
            if event.button == 1:
                click = True
        if start_button.collidepoint(mouse):
            if click:
                start_algo(algo)
        if reset_button.collidepoint(mouse):
            if click:
                numberOfLines = sizeoptions[option]
                height, colorOfLines = makearray(numberOfLines)

    # main display funcion
    
#     list1.draw(WIN)
    display()
    width = getwidth(height)
    
    # update the window 
    pygame.display.flip()
            

# end everything
pygame.quit()
    
    
```
